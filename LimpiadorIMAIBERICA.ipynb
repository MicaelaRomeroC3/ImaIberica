```python
import pandas as pd

# Ruta del archivo Excel en tu Google Drive
file_path = '/content/drive/MyDrive/Ima Iberica/BBDD_IMA_FEBRERO_2025 3.xlsx'

# Cargar el archivo Excel
df = pd.read_excel(file_path)

# Mostrar las primeras filas del archivo
df.head()
```

```python
# Limpiar los espacios al principio y al final de los nombres en la columna "Agente"
df['AGENTE'] = df['AGENTE'].str.strip()

# Verificar el resultado
df['AGENTE'].head()
```

# **COMPARAR AGENTES DEL LISTADO CON LOS DE LA BBDD**

```python
# Ruta del archivo Excel en tu Google Drive
file_path = '/content/drive/MyDrive/Ima Iberica/agentes a comparar.xlsx'

# Cargar el archivo Excel
Agentes_a_comparar = pd.read_excel(file_path)

# Mostrar las primeras filas del archivo
Agentes_a_comparar.head()
```

```python
# Comparar los agentes del archivo con la columna "AGENTE" de df
agentes_incorrectos = df[~df['AGENTE'].isin(Agentes_a_comparar['AGENTES_A_COMPARAR'])]

# Eliminar duplicados
agentes_incorrectos = agentes_incorrectos.drop_duplicates(subset=['AGENTE'])

# Mostrar los agentes incorrectos
pd.set_option('display.max_rows', None)  # Muestra todas las filas

print("Agentes con nombres incorrectos:")
print(agentes_incorrectos[['AGENTE']])
print(agentes_incorrectos[['AGENTE']].count())
```

# **AGENTES QUE YA NO ESTAN EN LA BBDD**

```python
# Lista de agentes que faltan en df['AGENTE']
agentes_no_en_bbdd = []

# Iterar sobre cada agente en Agentes_a_comparar['AGENTES_A_COMPARAR']
for agente in Agentes_a_comparar['AGENTES_A_COMPARAR']:
    # Si el agente no está en df['AGENTE'], agregarlo a la lista
    if agente not in df['AGENTE'].values:
        agentes_no_en_bbdd.append(agente)

# Mostrar los agentes que ya no están en la base de datos
print("\nAgentes que ya no están en la base de datos:")
print(agentes_no_en_bbdd)

# Verificar la cantidad de agentes no encontrados
print("Número de agentes no encontrados:", len(agentes_no_en_bbdd))
```
